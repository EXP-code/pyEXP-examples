# Scale coefficients by 43007.1 for Gadget units

Begin with the usual imports

```python
import os
import copy
import yaml
import time
import pyEXP
import numpy as np
from os.path import exists
```

### Read the coefficients

```python
coeffile = 'my_test.h5' # My test set...
coefs = pyEXP.coefs.Coefs.factory(coeffile)
```

### Now run through a loop and scale them by the desired factor and save the resulting set

```python
# This is the factor
factor = 43007.1

# This is the list of time keys
times = coefs.Times()

for time in times:
    coef = coefs.getCoefStruct(time)
    coef.coefs *= factor
    
# Now save the scaled co
```
